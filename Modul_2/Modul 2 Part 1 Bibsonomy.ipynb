```python
items = data['items']
print(type(items))
print(len(items))
```

```python
publications = [item for item in items if item['type'] == 'Publication']
print(len(publications))
```

```python
pubs_with_abstracts = [item for item in publications if 'abstract' in item]
print(len(pubs_with_abstracts))
```
