```python
path = 'rohstoffe.csv'
```

```python
with open(path) as csvfile:
    lines = csvfile.readlines()
    
header = lines[0].strip().split(',')
print(header)
```

```python
first_line = lines[3].strip().split(',')

for i in range(len(first_line)):
    print(header[i], ':', first_line[i])
```

```python
import csv
```

```python
data = []

with open(path, 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    for row in reader:
        data.append(row)
```

```python
header = data[0]
print(header)
```

```python
first_line = data[3]
for i in range(len(first_line)):
    print(header[i], ':', first_line[i])
```

```python
with open(path, 'r') as csvfile:
    reader = csv.DictReader(csvfile, delimiter=',')
    for row in reader:
        print("Title: {}\nAuthors: {}\nGenre: {}\n".format(row['Title'], row['Author'], row['Custom3']))
```
